# Contribution Plotting Utility
The purpose of this workbook is to demonstrate the different ways to plot integral index contributions. First, we compute the contributions to the integral index E for a specific test case

```python
from tsunami_ip_utils.integral_indices import calculate_E_contributions, calculate_E
from tsunami_ip_utils.viz import plot_contributions
from uncertainties import ufloat
import matplotlib.pyplot as plt

# List of all SDF files
all_sdfs = [ f"3d-sphere/sphere_model_{i}.sdf" for i in range(1, 13) ]
E = calculate_E(all_sdfs[0:1], all_sdfs[1:2])
E = E[0,0]
print(f"The total integral index is E = {E:1.4f}")
E_contributions_nuclide, E_contributions_nuclide_reaction = calculate_E_contributions(all_sdfs[0:1], all_sdfs[1:2])
```

## Nuclide-wise and Nuclide-Reaction-wise contributions

### Nuclide-wise

#### Bar chart

```python
test_case = E_contributions_nuclide[0][0]
plot_contributions(test_case, 'bar', 'E')
```

#### Pie chart

```python
plot_contributions(test_case, 'pie', 'E')
```

#### Interactive Pie Chart

```python
fig = plot_contributions(test_case, 'interactive_pie', 'E', interactive_legend=False)
fig.show()
```

### Nuclide-Reaction-Wise Contributions

#### Bar Chart

```python
test_case = E_contributions_nuclide_reaction[0][0]
plot_contributions(test_case, 'bar', 'E')
```

#### Pie Chart

```python
plot_contributions(test_case, 'pie', 'E')
```

#### Interactive Pie Chart (with interactive legend)

```python
fig = plot_contributions(test_case, 'interactive_pie', 'E')
fig.show()
```
